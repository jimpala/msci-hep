```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
```

```python
df_2jet2fold = pd.DataFrame.from_csv('gs_BDT_2jet2fold.csv')
df_2jet2fold
```

```python
x = range(100)
y = range(100,200)
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x[:4], y[:4], s=10, c='b', marker="s", label='first')
ax1.scatter(x[40:],y[40:], s=10, c='r', marker="o", label='second')
plt.legend(loc='upper left');
plt.show()
```

```python
scores_2depth = df_2jet2fold[df_2jet2fold['param_base_estimator__max_depth'] == 2]['mean_train_score'].as_matrix()
scores_3depth = df_2jet2fold[df_2jet2fold['param_base_estimator__max_depth'] == 3]['mean_train_score'].as_matrix()
scores_4depth = df_2jet2fold[df_2jet2fold['param_base_estimator__max_depth'] == 4]['mean_train_score'].as_matrix()

scores = np.vstack((scores_4depth, scores_3depth, scores_2depth))
scores
```

```python
fig, ax = plt.subplots(figsize=(15,3))
cmap2 = sns.light_palette("green", as_cmap=True)

# rcParams
mpl.rcParams['axes.titlesize'] = 16.
mpl.rcParams['axes.titleweight'] = 'bold'


ax.imshow(scores, cmap=cmap2,
          interpolation='none', extent=[100,410,2,5], aspect='auto')
plt.xticks(np.arange(100, 411, 10))
plt.yticks(np.arange(2, 5.1, 1))

# Hide major tick labels
ax.set_xticklabels('')
ax.set_yticklabels('')


# Customize minor tick labels
ax.set_xticks(np.arange(105, 415, 10), minor=True)
ax.set_xticklabels(np.arange(100, 410, 10), minor=True)

# Customize minor tick labels
ax.set_yticks([2.5,3.5,4.5],      minor=True)
ax.set_yticklabels(['2','3','4'], minor=True)

# Titles
ax.set_xlabel('Number of trees')
ax.set_ylabel('Maximum tree depth')
ax.set_title('Scikit BDT Hyperparameter Scan', loc='left')

    
plt.show()
```
